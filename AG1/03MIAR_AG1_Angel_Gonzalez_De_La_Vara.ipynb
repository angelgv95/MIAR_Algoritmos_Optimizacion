## Devolución de cambio por técnica voraz

```python
def cambio_monedas(N, SM):
  SOLUCION = [0]*len(SM)   #SOLUCION = [0,0,0,0,..]
  ValorAcumulado = 0

  for i,valor in enumerate(SM):
    monedas =  (N-ValorAcumulado)//valor
    SOLUCION[i] = monedas
    ValorAcumulado = ValorAcumulado + monedas*valor

    if ValorAcumulado == N:
      return SOLUCION


cambio_monedas(15,[25,10,5,1])
```

## N-Reinas por técnica de vueta atrás


```python
def escribe(S):
  n = len(S)
  for x in range(n):
    print("")
    for i in range(n):
      if S[i] == x+1:
        print(" X " , end="")
      else:
        print(" - ", end="")


def es_prometedora(SOLUCION,etapa):
  #print(SOLUCION)
  #Si la solución tiene dos valores iguales no es valida => Dos reinas en la misma fila
  for i in range(etapa+1):
    #print("El valor " + str(SOLUCION[i]) + " está " +  str(SOLUCION.count(SOLUCION[i])) + " veces")
    if SOLUCION.count(SOLUCION[i]) > 1:
      return False

    #Verifica las diagonales
    for j in range(i+1, etapa +1 ):
      #print("Comprobando diagonal de " + str(i) + " y " + str(j))
      if abs(i-j) == abs(SOLUCION[i]-SOLUCION[j]) : return False
  return True



def reinas(N, solucion=[], etapa=0):
  if len(solucion) == 0:
      solucion=[0 for i in range(N)]

  for i in range(1, N+1):
    solucion[etapa] = i

    if es_prometedora(solucion, etapa):
      if etapa == N-1:
        print(solucion)
        escribe(solucion)
        print()
      else:
        # escribe(solucion)
        # print()
        reinas(N, solucion, etapa+1)
    else:
      None

    solucion[etapa] = 0

reinas(4)
```

## Viaje por el rio. Programación dinámica

```python
TARIFAS = [
[0,5,4,3,999,999,999],
[999,0,999,2,3,999,11],
[999,999, 0,1,999,4,10],
[999,999,999, 0,5,6,9],
[999,999, 999,999,0,999,4],
[999,999, 999,999,999,0,3],
[999,999,999,999,999,999,0]
]



################################################################
def Precios(TARIFAS):
################################################################
  #Total de Nodos
  N = len(TARIFAS[0])

  #Inicialización de la tabla de precios
  PRECIOS = [ [9999]*N for i in [9999]*N]
  RUTA = [ [""]*N for i in [""]*N]

  for i in range(0,N-1):
    RUTA[i][i] = i             #Para ir de i a i se "pasa por i"
    PRECIOS[i][i] = 0          #Para ir de i a i se se paga 0
    for j in range(i+1, N):
      MIN = TARIFAS[i][j]
      RUTA[i][j] = i

      for k in range(i, j):
        if PRECIOS[i][k] + TARIFAS[k][j] < MIN:
            MIN = min(MIN, PRECIOS[i][k] + TARIFAS[k][j] )
            RUTA[i][j] = k          #Anota que para ir de i a j hay que pasar por k
        PRECIOS[i][j] = MIN

  return PRECIOS,RUTA
################################################################


PRECIOS,RUTA = Precios(TARIFAS)
#print(PRECIOS[0][6])

print("PRECIOS")
for i in range(len(TARIFAS)):
  print(PRECIOS[i])

print("\nRUTA")
for i in range(len(TARIFAS)):
  print(RUTA[i])

#Determinar la ruta con Recursividad
def calcular_ruta(RUTA, desde, hasta):
  if desde == hasta:
    #print("Ir a :" + str(desde))
    return ""
  else:
    return str(calcular_ruta( RUTA, desde, RUTA[desde][hasta])) +  \
                ',' + \
                str(RUTA[desde][hasta] \
              )

print("\nLa ruta es:")
calcular_ruta(RUTA, 0,6)
```

## Actividad extra
Problema: Encontrar los dos puntos más cercanos
- Dado un conjunto de puntos se trata de encontrar los dos puntos más cercanos
- Guía para aprendizaje:
- Suponer en 1D, o sea, una lista de números: [3403, 4537, 9089, 9746, 7259, ….
- Primer intento: Fuerza bruta
- Calcular la complejidad. ¿Se puede mejorar?
- Segundo intento. Aplicar Divide y Vencerás
- Calcular la complejidad. ¿Se puede mejorar?
- Extender el algoritmo a 2D: [(1122, 6175), (135, 4076), (7296, 2741)…
- Extender el algoritmo a 3D.

### Primer intento: fuerza bruta

```python
import random

# Creamos una función para calcular el par más cercano en una lista 1D por fuerza bruta
def par_mas_cercano_1D_fb(lst):

    # Inicializamos la distancia mínima a infinito y creamos una lista vacía que contendrá el par de números más
    # cercanos entre sí
    dist_min = float('inf')
    par_cercanos = []

    # Recorremos la lista y comparamos cada valor con los demás. Solo es necesario comparar un valor con los siguientes
    # en la lista, ya que los anteriores ya han sido comparados.
    for i, n1 in enumerate(lst[0:-1]):
        for n2 in lst[i+1:]:

            # Calculamos la distancia entre los dos puntos en valor absoluto
            dist = abs(n2 - n1)

            # Si la distancia entre estos dos puntos es más pequeña que la distancia más pequeña ya vista anteriormente,
            # definimos este nuevo par de puntos como los más cercanos
            if dist < dist_min:
                dist_min = dist
                par_cercanos = [n1, n2]

    return par_cercanos, dist_min

# Generamos la lista de números aleatorios
lista_1D = [random.randrange(1, 10000000) for x in range(1000)]

# LLamamos a la función y mostramos el resultado por pantalla
par_cercanos, dist_min = par_mas_cercano_1D_fb(lista_1D)
print(f'El par de puntos más cercano es {par_cercanos}, a una distancia de {dist_min}')

```

El algoritmo de fuerza bruta recorre la lista comparando todos los pared posibles. Esto implica n(n-1)/2 comparaciones, por lo que **la complejidad del algoritmo es O(n^2)**.

La complejidad del algoritmo por fuerza bruta se puede mejorar, por ejemplo, ordenando la lista antes de lanzar el algoritmo de comparación por fuerza bruta. De esta forma solo hay que comparar los elementos consecutivos.

La complejidad de un algoritmo de ordenación eficaz es de O(n*log(n)) en el peor caso(como quick sort), mientras que la complejidad de comparar la distancia entre valores consecutivos en una lista ordenada es de O(n), ya que la lista solo se tiene que recorrer una vez. Por lo tanto la complejidad total sería de:

O(n*log(n)) + O(n) = O(n*log(n))

### Segundo intento: Divide y vencerás

```python
# En primer lugar se define una función recursiva que compara la distancia de los elementos de una lista ordenada,
# diviendo esta lista es sublistas más pequeñas de forma recursiva
def recursivo_mas_cercano(lst):

    # Comprobamos la dimensión de la lista. Si solo hay dos elementos, se devuelven esos elementos y la distancia entre
    # ellos.
    # Si solo hay un elemento, la distancia se deja en infinito al no poder calcularse
    n = len(lst)
    if n == 2:
        return [lst[0], lst[1]], abs(lst[1] - lst[0])
    elif n == 1:
        return None, float('inf')

    # Se divide la lista en dos mitades y se llama de forma recursiva la función en cada una de las dos mitades
    # Al estar la lista ordenada, esto permite poder comparar la distancia entre los elementos contiguos fácilmente
    mitad = n // 2
    par1, dist1 = recursivo_mas_cercano(lst[0:mitad])
    par2, dist2 = recursivo_mas_cercano(lst[mitad:])

    # Se compara la distancia mínima obtenida en las dos mitades de la lista y se toma la más pequeña y el par de
    # puntos correspondiente
    if dist2 < dist1:
        dist_min = dist2
        par_cercanos = par2
    else:
        dist_min = dist1
        par_cercanos = par1

    # Es necesario comparar también la distancia entre los extremos interiores de las dos sublistas
    # Por ejemplo, si tenemos las dos sublistas [1, 3] y [4, 8], es necesario comprobar la distancia entre 3 y 4
    dist_extremos = abs(lst[mitad] - lst[mitad - 1])

    # Si la distancia entre los extremos interiores es menor que la tomada anteriormente, se actualiza la distancia
    # mínima y el par de puntos correspondiente
    if dist_extremos < dist_min:
        dist_min = dist_extremos
        par_cercanos = [lst[mitad - 1], lst[mitad]]

    # Finalmente se devuelven los resultados
    return par_cercanos, dist_min


# Creamos la función global para calcular el par más cercano en una lista 1D mediante divide y vencerás
def par_mas_cercano_1D_dv(lst):

    # En primer lugar ordenamos la lista
    lst = sorted(lst)

    # Finalmente llamamos a la función recursiva que devuelve el par de puntos más cercanos mediante divide y vencerás
    return recursivo_mas_cercano(lst)


# LLamamos a la función y mostramos el resultado por pantalla
par_cercanos, dist_min = par_mas_cercano_1D_dv(lista_1D)
print(f'El par de puntos más cercano es {par_cercanos}, a una distancia de {dist_min}')
```

Este algoritmo se divide en dos fases: primero se ordena la lista de puntos en orden ascendente y después de aplica el algoritmo de divide y vencerás de forma recursiva para encontrar los puntos más cercanos.

El algoritmo de ordenación usado es la función sorted() de python, que se basa en el algoritmo Timsort y que tiene una complejidad O(n*log(n)) en el peor de los casos.

El algoritmo recursivo de comparación de distancias mediante divide y vencerás también tiene una complejidad de O(n*log(n)).

Por lo tanto, **la complejidad total del algoritmo completo mediante divide y vencerás también es de O(n*log(n))**. Además, este orden de complejidad **no se puede mejorar** ya que viene dimensionado por la complejidad del algoritmo de ordenación.

Nótese que este algoritmo presenta la misma complejidad que la mejora propuesta en el ejercicio anterior para el algoritmo de fuerza bruta (primero ordenación y después comparación directa de los valores contiguos). El algoritmo de divide y vencerás muestra sus beneficios en el caso 2D.

## Extensión del algoritmo de divide y vencerás a 2D

```python
import math
import random

# En primer lugar creamos una función que nos ayuda a calcular la distancia entre dos puntos en un espacio 2D
def distancia(p1, p2):
    return math.sqrt((p1[0] - p2[0])**2 + (p1[1] - p2[1])**2)

# Creamos una función recursiva que calcula el par de puntos más cercano en un conjunto 2D
def recursivo_mas_cercano_2D(px, py):

    # Obtenemos el número de puntos
    n = len(px)

    # Si hay 3 o menos puntos, se calcula directamente la distancia entre todos los pares posibles devolviendo el par
    # con menor distancia
    if n <= 3:
        mejor_par = None
        mejor_dist = float('inf')
        for i in range(n):
            for j in range(i + 1, n):
                d = distancia(px[i], px[j])
                if d < mejor_dist:
                    mejor_dist = d
                    mejor_par = (px[i], px[j])
        return mejor_par, mejor_dist

    # Dividimos el conjunto de puntos en dos mitades respecto a la coordenada x
    mitad = n // 2
    x_medio = px[mitad][0]

    # Separamos las listas en subconjuntos izquierdo y derecho
    px_izq = px[:mitad]
    px_der = px[mitad:]

    # Construimos las listas ordenadas por y correspondientes a cada mitad
    py_izq = [p for p in py if p[0] <= x_medio]
    py_der = [p for p in py if p[0] > x_medio]

    # Llamamos de forma recursiva a cada mitad para obtener el mejor par en cada subconjunto
    par1, dist1 = recursivo_mas_cercano_2D(px_izq, py_izq)
    par2, dist2 = recursivo_mas_cercano_2D(px_der, py_der)

    # Comparamos las mejores distancias de ambas mitades y seleccionamos la menor
    if dist1 < dist2:
        mejor_par, mejor_dist = par1, dist1
    else:
        mejor_par, mejor_dist = par2, dist2

    # Construimos la franja central que contiene los puntos cuya coordenada x está a una distancia menor que mejor_dist
    # de la línea media
    franja = [p for p in py if abs(p[0] - x_medio) < mejor_dist]

    # Comparamos cada punto de la franja con los siguientes (como máximo 7)
    # Esto se basa en propiedades geométricas que limitan el número de comparaciones necesarias
    for i in range(len(franja)):
        for j in range(i + 1, min(i + 8, len(franja))):
            d = distancia(franja[i], franja[j])
            if d < mejor_dist:
                mejor_dist = d
                mejor_par = (franja[i], franja[j])

    # Finalmente se devuelve el mejor par encontrado y su distancia
    return mejor_par, mejor_dist

# Creamos la función global para calcular el par más cercano en una lista 2D mediante divide y vencerás
def par_mas_cercano_2D_dv(puntos):

    # Creamos dos listas ordenadas, la primera en función de la coordenada x y la segunda en función de la coordenada y
    px = sorted(puntos, key=lambda p: p[0])
    py = sorted(puntos, key=lambda p: p[1])

    # Finalmente llamamos a la función recursiva que devuelve el par de puntos más cercanos mediante divide y vencerás
    return recursivo_mas_cercano_2D(px, py)

# Definimos el dataset de test con puntos en 2D
lista_2D = [(random.randrange(1, 10000), random.randrange(1, 10000)) for x in range(1000)]

# LLamamos a la función y mostramos el resultado por pantalla
par_cercanos, dist_min = par_mas_cercano_2D_dv(lista_2D)
print(f'El par de puntos más cercano es {par_cercanos}, a una distancia de {dist_min}')
```

## Extensión del algoritmo de divide y vencerás a 3D

```python
import math
import random

# En primer lugar creamos una función que nos ayuda a calcular la distancia entre dos puntos en un espacio 3D
def distancia_3D(p1, p2):
    return math.sqrt(
        (p1[0] - p2[0])**2 +
        (p1[1] - p2[1])**2 +
        (p1[2] - p2[2])**2
    )

# Creamos una función auxiliar que calcula el par de puntos más cercano por fuerza bruta
def fuerza_bruta_3D(puntos):
    mejor_par = None
    mejor_dist = float('inf')

    # Comparamos todos los pares posibles y nos quedamos con el de menor distancia
    n = len(puntos)
    for i in range(n):
        for j in range(i + 1, n):
            d = distancia_3D(puntos[i], puntos[j])
            if d < mejor_dist:
                mejor_dist = d
                mejor_par = (puntos[i], puntos[j])

    return mejor_par, mejor_dist

# Creamos una función recursiva que calcula el par de puntos más cercano en un conjunto 3D
def recursivo_mas_cercano_3D(px):

    # Obtenemos el número de puntos
    n = len(px)

    # Si hay 3 o menos puntos, se resuelve directamente por fuerza bruta
    if n <= 3:
        return fuerza_bruta_3D(px)

    # Dividimos el conjunto de puntos en dos mitades respecto a la coordenada x
    mitad = n // 2
    x_medio = px[mitad][0]

    # Separamos la lista en los subconjuntos izquierdo y derecho
    px_izq = px[:mitad]
    px_der = px[mitad:]

    # Llamamos de forma recursiva a cada mitad para obtener el mejor par en cada subconjunto
    par1, dist1 = recursivo_mas_cercano_3D(px_izq)
    par2, dist2 = recursivo_mas_cercano_3D(px_der)

    # Comparamos las mejores distancias de ambas mitades y seleccionamos la menor
    if dist1 < dist2:
        mejor_par = par1
        mejor_dist = dist1
    else:
        mejor_par = par2
        mejor_dist = dist2

    # Construimos la franja central que contiene los puntos cuya coordenada x está a una distancia menor que mejor_dist
    # de la línea media
    franja = [p for p in px if abs(p[0] - x_medio) < mejor_dist]

    # Comparamos todos los pares posibles dentro de la franja
    # En 3D no existe una cota fija equivalente a los 7 vecinos del caso 2D, por lo que se comparan todos los pares
    # dentro de la franja
    m = len(franja)
    for i in range(m):
        for j in range(i + 1, m):
            d = distancia_3D(franja[i], franja[j])
            if d < mejor_dist:
                mejor_dist = d
                mejor_par = (franja[i], franja[j])

    # Finalmente se devuelve el mejor par encontrado y su distancia
    return mejor_par, mejor_dist

# Creamos la función global para calcular el par más cercano en una lista 3D mediante divide y vencerás
def par_mas_cercano_3D_dv(puntos):

    # En primer lugar ordenamos los puntos por coordenada x
    px = sorted(puntos, key=lambda p: p[0])

    # Finalmente llamamos a la función recursiva que devuelve el par de puntos más cercanos
    return recursivo_mas_cercano_3D(px)

# Definimos el dataset de test con puntos en 3D
lista_3D = [(random.randrange(1, 10000), random.randrange(1, 10000), random.randrange(1, 10000)) for x in range(1000)]

# LLamamos a la función y mostramos el resultado por pantalla
par_cercanos, dist_min = par_mas_cercano_3D_dv(lista_3D)
print(f'El par de puntos más cercano es {par_cercanos}, a una distancia de {dist_min}')
```
